# Feature Scaling:- 
Feature Scaling is a technique to standardize the independent features present in the data in a fixed range. 
It is performed during the data pre-processing to handle highly varying magnitudes or values or units.
In other words it is used to range the data from (-1 to +1), (0 to 1).
It is required because we have a lot of disperity bertween extreme values and our data whichmeans data is not normalized.
So thats why we dont get the desired output.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
x = np.array([[-400],[-100],[0],[100],[400]])
```

```python
x
```

# MinMaxScaler :- 
It shrinks the range such that it is now between 0 and 1 (or -1 to 1 if there exist negative values). 
The MinMaxScaler works well for cases when the distribution is not Gaussian or when the standard deviation is very small.

#### x(i) - min(x) / max(x) - min(x)

##### Applying on Array

```python
minmaxscaler = preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
#Now fit the data
x_scaler = minmaxscaler.fit_transform(x)
```

```python
print(x_scaler)
```

```python
# (-400 - (-400)) / (400 - (-400)) = 0
# (-100 - (-400)) / (400 - (-400)) = 3/8 = 0.375
# And so on...
```

# MaxAbsScaler :- 
The MaxAbsScaler works very similarly to the MinMaxScaler but automatically scales the data to a [-1,1] range based on the absolute maximum. 
This scaler is meant for data that is already centered at zero or sparse data. 
It does not shift/center the data, and thus does not destroy any sparsity.

```python
x = np.array([[-500.5], 
              [-100.1], 
              [0], 
              [100.1], 
              [900.9]])
```

# Robust Scaler :-  
Scale features using statistics that are robust to outliers.

This Scaler removes the median and scales the data according to the Interquartile Range (IQR). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile)

##### (Xi - Q2) / (Q3 - Q1)
Q1 -> Upper quartile
Q2 -> Median
Q3 -> Lower quartile

```python
x = np.array([[-500], 
              [-100], 
              [0], 
              [100], 
              [900]])
```

```python
x
```

```python
robustscale = preprocessing.RobustScaler()
```

```python
x_robust = robustscale.fit_transform(x)
```

```python
x_robust
```

```python
# To check whether the upper answer is correct or not
```

```python
q1 = np.percentile(x,25)
q2 = np.percentile(x,50)
q3 = np.percentile(x,75)
```

```python
print(q1)
print(q2)
print(q3)
```
